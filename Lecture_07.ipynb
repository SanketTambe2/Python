```python
## Lecture - 07

#List
'''
List
Use: Store multiple items.
Syntax: Square brackets []
Changeable (Mutable): Yes

Example:
my_list = [1, 2, 3, 4]

'''

##List Functions / Methods

'''ðŸ‘‰

append() â†’ add item at end
l = [1,2]
l.append(3)
print(l)   # [1,2,3] Outputs


insert() â†’ add item at position
l.insert(1, 5)
print(l)   # [1,5,2,3]


extend() â†’ add multiple items
l.extend([7,8])
print(l)   # [1,5,2,3,7,8]


remove() â†’ remove first match
l.remove(5)
print(l)   # [1,2,3,7,8]


pop() â†’ remove by index (default last)
l.pop()
print(l)   # [1,2,3,7]


clear() â†’ remove all items
l.clear()
print(l)   # []


index() â†’ find index of value
l = [10,20,30]
print(l.index(20))  # 1


count() â†’ count how many times value appears
l = [1,1,2,3]
print(l.count(1))   # 2


sort() â†’ sort list
l = [3,1,2]
l.sort()
print(l)   # [1,2,3]


reverse() â†’ reverse order
l.reverse()
print(l)   # [3,2,1]


copy() â†’ copy list
l2 = l.copy()
print(l2)  # [3,2,1]

'''

## Convert List â†’ Tuple

'''ðŸ‘‰ Use tuple()

l = [1, 2, 3]
t = tuple(l)
print(t)   # (1, 2, 3)

'''

## Convert List â†’ Set

'''ðŸ‘‰ Use set()

l = [1, 2, 2, 3]
s = set(l)
print(s)   # {1, 2, 3}   # duplicates removed

'''

## Convert List â†’ Dictionary

'''ðŸ‘‰ List must have pairs (key, value).

l = [["a", 1], ["b", 2]]
d = dict(l)
print(d)   # {'a': 1, 'b': 2}

'''


## Program

l1 = [11, 22, 33, 44, 55, 66, 77]   # list

for i in l1:    # directly loop through list
    print(i)

```

```python
## Tuples

'''
Tuple
Use: Store multiple items.
Syntax: Round brackets ()
Changeable (Mutable): No

Example:
my_tuple = (1, 2, 3, 4)

'''

## Tuple Methods (only 2 main ones)

'''

count() â†’ counts how many times a value appears
t = (1, 2, 2, 3)
print(t.count(2))   # 2 Outputs


index() â†’ gives position of first occurrence
t = (10, 20, 30)
print(t.index(20))  # 1

'''

## Convert Tuple â†’ List

'''
ðŸ‘‰ Use list()

t = (1, 2, 3)
l = list(t)
print(l)   # [1, 2, 3]

'''

## Convert Tuple â†’ Set

'''
ðŸ‘‰ Use set()

t = (1, 2, 3, 2)
s = set(t)
print(s)   # {1, 2, 3}   (removes duplicates)

'''

## Convert Tuple â†’ Dictionary

'''

ðŸ‘‰ Tuple must have pairs (key, value).

t = (("a", 1), ("b", 2))
d = dict(t)
print(d)   # {'a': 1, 'b': 2}

'''


## Program

t1 = (11, 22, 33, 44, 55, 66, 77)   # tuple
l = len(t1)                         # length of tuple
x = 0

while x != l:                       # loop until x reaches length
    print(t1[x])                    # print element at index x
    x = x + 1                          # increase x


'''

t1 â†’ tuple with numbers.
len(t1) â†’ gives total elements (7).
x = 0 â†’ loop starts from index 0.
while x != l: â†’ loop runs until x becomes 7.
print(t1[x]) â†’ prints each element.
x = x + 1 â†’ moves to next index.

'''
```

```python
## Slicing Operator [:]

'''

Used to extract part of a sequence (list, tuple, string).
Syntax â†’ sequence[start : stop : step]
start â†’ index where slice begins (default = 0).
stop â†’ index where slice ends (not included).
step â†’ how many to skip (default = 1).

'''

## Examples with List

'''

nums = [10, 20, 30, 40, 50]

print(nums[1:4])     # [20, 30, 40]         (from index 1 to 3)
print(nums[:3])      # [10, 20, 30]         (from start to index 2)
print(nums[2:])      # [30, 40, 50]         (from index 2 to end)
print(nums[::2])     # [10, 30, 50]         (step 2, skip every second)
print(nums[::-1])    # [50, 40, 30, 20, 10] (reverse list)

'''
```
